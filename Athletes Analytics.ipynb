```python
#Assign URL for Best by Athlete's Records from Nigeria
#Men
best_men100m = "https://worldathletics.org/records/toplists/sprints/100-metres/outdoor/men/senior/2023?regionType=countries&region=ngr&timing=electronic&windReading=regular&page=1&bestResultsOnly=true"
best_men200m = "https://worldathletics.org/records/toplists/sprints/200-metres/outdoor/men/senior/2023?regionType=countries&region=ngr&timing=electronic&windReading=regular&page=1&bestResultsOnly=true"
best_men400m = "https://worldathletics.org/records/toplists/sprints/400-metres/outdoor/men/senior/2023?regionType=countries&region=ngr&timing=electronic&page=1&bestResultsOnly=true"
best_men110mH = "https://worldathletics.org/records/toplists/hurdles/110-metres-hurdles/outdoor/men/senior/2023?regionType=countries&region=ngr&timing=electronic&windReading=regular&page=1&bestResultsOnly=true"
best_men400mH = "https://worldathletics.org/records/toplists/hurdles/400-metres-hurdles/outdoor/men/senior/2023?regionType=countries&region=ngr&timing=electronic&page=1&bestResultsOnly=true"

#Women
best_women100m = "https://worldathletics.org/records/toplists/sprints/100-metres/outdoor/women/senior/2023?regionType=countries&region=ngr&timing=electronic&windReading=regular&page=1&bestResultsOnly=true"
best_women200m = "https://worldathletics.org/records/toplists/sprints/200-metres/outdoor/women/senior/2023?regionType=countries&region=ngr&timing=electronic&windReading=regular&page=1&bestResultsOnly=true"
best_women400m = "https://worldathletics.org/records/toplists/sprints/400-metres/outdoor/women/senior/2023?regionType=countries&region=ngr&timing=electronic&page=1&bestResultsOnly=true"
best_women100mH = "https://worldathletics.org/records/toplists/hurdles/100-metres-hurdles/outdoor/women/senior/2023?regionType=countries&region=ngr&timing=electronic&windReading=regular&page=1&bestResultsOnly=true"
best_women400mH = "https://worldathletics.org/records/toplists/hurdles/400-metres-hurdles/outdoor/women/senior/2023?regionType=countries&region=ngr&timing=electronic&page=1&bestResultsOnly=true"
```

```python
#Create a list of all the events in each category
all_events = [all_men100m, all_men200m, all_men400m, all_men110mH, all_men400mH, all_women100m, all_women200m, all_women400m, all_women100mH, all_women400mH]

best_events = [best_men100m, best_men200m, best_men400m, best_men110mH, best_men400mH, best_women100m, best_women200m, best_women400m, best_women100mH, best_women400mH]
```

```python
# Initialize an empty list to store the Events in a DataFrame
all_event_table = []
best_event_table = []

# Use a for loop to read each events and store them in the DataFrames
for event in all_events:
    df = pd.read_html(event)[0]
    all_event_table.append(df)
    
for event in best_events:
    df = pd.read_html(event)[0]
    best_event_table.append(df)
```

```python
#Merge all the dataframes in the event_tables into a single dataframe
all_events_performance = pd.concat(all_event_table, ignore_index=True)
best_events_performance = pd.concat(best_event_table, ignore_index=True)
```

```python
#Remove Unwanted Columns
all_events_performance = all_events_performance.drop(columns = ["Rank", "Unnamed: 6", "Unnamed: 7"])
best_events_performance = best_events_performance.drop(columns = ["Rank", "Unnamed: 6", "Unnamed: 7"])
```

```python
all_events_performance = all_events_performance.sort_values(by='Results Score', ascending=False)
best_events_performance = best_events_performance.sort_values(by='Results Score', ascending=False)
```

```python
all_events_performance = all_events_performance.reset_index().drop(columns = "index")
best_events_performance = best_events_performance.reset_index().drop(columns = "index")
```

```python
#Start Indexing from 1
all_events_performance.index = all_events_performance.index + 1
best_events_performance.index = best_events_performance.index + 1
```

```python
#Rename Index Column
all_events_performance = all_events_performance.rename_axis("Ranking")
best_events_performance = best_events_performance.rename_axis("Ranking")
```
