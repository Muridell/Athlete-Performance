```python
#Create a list of all the events in each category
all_events = [all_men100m, all_men200m, all_men400m, all_men110mH, all_men400mH, all_women100m, all_women200m, all_women400m, all_women100mH, all_women400mH]

best_events = [best_men100m, best_men200m, best_men400m, best_men110mH, best_men400mH, best_women100m, best_women200m, best_women400m, best_women100mH, best_women400mH]

event_name = ["Men's 100m", "Men's 200m", "Men's 400m", "Men's 110mH", "Men's 400mH", "Women's 100m", "Women's 200m", "Women's 400m", "Women's 100mH", "Women's 400mH"]
```

```python
# Initialize an empty list to store the Events in a DataFrame
all_event_table = []
best_event_table = []

# Use a for loop to read each events and store them in the DataFrames
for event in all_events:
    df = pd.read_html(event)[0]
    all_event_table.append(df)
    
for event in best_events:
    df = pd.read_html(event)[0]
    best_event_table.append(df)
```

```python
#Add Event Name to Each Event in the DataFrames Category

for i in range(len(all_event_table)):
    all_event_table[i]['Event'] = event_name[i]
    
for i in range(len(best_event_table)):
    best_event_table[i]['Event'] = event_name[i]
```

```python
#Merge all the dataframes in the event_tables into a single dataframe
all_events_performance = pd.concat(all_event_table, ignore_index=True)
best_events_performance = pd.concat(best_event_table, ignore_index=True)
```

```python
#Remove Unwanted Columns
all_events_performance = all_events_performance.drop(columns = ["Rank", "Unnamed: 6", "Unnamed: 7"])
best_events_performance = best_events_performance.drop(columns = ["Rank", "Unnamed: 6", "Unnamed: 7"])
```

```python
#Sort all data by Top Performance Record
all_events_performance = all_events_performance.sort_values(by='Results Score', ascending=False)
best_events_performance = best_events_performance.sort_values(by='Results Score', ascending=False)
```
