```python
# Initialize an empty list to store the Events in a DataFrame
all_events_table = []

# Use a for loop to read each events and store them in the DataFrames
for event in events:
    df = pd.read_html(event)[0]
    all_events_table.append(df)
```

```python
#Add Event Name to Each Event in the DataFrames Category

for i in range(len(all_events_table)):
    all_events_table[i]['Event'] = events_name[i]
```

```python
#Merge all the dataframes in the event_tables into a single dataframe
all_events_performance = pd.concat(all_events_table, ignore_index=True)
all_events_performance
```

```python
#Create a New table from the Main Table by selecting just the needed columns
events_performance = all_events_performance[["Place","Name", "Nat.", "Event", "Mark"]]
events_performance
```

```python
#Rename the Columns in the New Table
events_performance = events_performance.rename(columns={'Place': 'Position', 'Nat.': 'Country'})
events_performance
```

```python
#Create an Athlete column by spliting the Name column using ,
events_performance["Athlete"] = events_performance["Name"].str.split(', ')
events_performance
```

```python
#Expand the Table by making each Athlete record in a row
events_performance = events_performance.explode("Athlete")
events_performance
```

```python
#Extract Country Name from Athlete Name
country_names = [
    "United States",
    "Italy",
    "Jamaica",
    "Great Britain & NI",
    "Japan",
    "France",
    "South Africa",
    "Brazil",
    "United States",
    "France",
    "Great Britain & NI",
    "Jamaica",
    "India",
    "Netherlands",
    "Italy",
    "Botswana",
    "United States",
    "Jamaica",
    "Great Britain & NI",
    "Italy",
    "Poland",
    "Germany",
    "Netherlands",
    "Côte d’Ivoire",
    "Switzerland",
    "Netherlands",
    "Jamaica",
    "Great Britain & NI",
    "Canada",
    "Belgium",
    "Poland",
    "Italy",
    "Ireland",
    "France",
    "United States",
    "Great Britain & NI",
    "Czech Republic",
    "France",
    "Belgium",
    "Ireland",
    "Germany",
    "Poland",
    "Netherlands",
]
```

```python
# Remove country names from athlete names using regex
for country in country_names:
    events_performance['Athlete'] = events_performance['Athlete'].str.replace(country, '')

events_performance
```

```python
events_performance = events_performance[["Position", "Athlete", "Country", "Event", "Mark"]]
events_performance
```

```python
# Define ranking points for individual events and relay events
individual_ranking_points = {1.0: 20, 2.0: 16, 3.0: 12, 4.0: 10, 5.0: 8, 6.0: 6, 7.0: 4, 8.0: 2}
relay_ranking_points = {1.0: 40, 2.0: 32, 3.0: 24, 4.0: 20, 5.0: 16, 6.0: 12, 7.0: 8, 8.0: 4}


# Create a function to Calculate ranking points based on event type and position
def calculate_ranking_points(row):
    event_type = row['Event']
    position = row['Position']
    
    if "Relay" in event_type:
        return relay_ranking_points.get(position, 0) * 0.25
    else:
        # Individual Event
        return individual_ranking_points.get(position, 0)

    
# Apply the calculate_ranking_points function to each row in the DataFrame
events_performance['Ranking Point'] = events_performance.apply(calculate_ranking_points, axis=1)

events_performance
```

```python
#Perform Athletes Ranking based Ranking Point
athletes_ranking = events_performance.groupby(["Athlete", "Country"])["Ranking Point"].sum().reset_index()
athletes_ranking = athletes_ranking.sort_values(by = "Ranking Point", ascending= False)
athletes_ranking
```

```python
# Create Athlete Ranking and handle ties
athletes_ranking['Ranking'] = athletes_ranking['Ranking Point'].rank(method='min', ascending=False)
athletes_ranking['Ranking'] = athletes_ranking['Ranking'].astype(int)

# Use the Ranking column as index
athletes_ranking = athletes_ranking.set_index('Ranking')
athletes_ranking
```

```python
#PERFORM COUNTRY RANKING FOR BASE ON SPECIFIC EVENTS
country_specific_events = [
    "Men's 100 Meters", 
    "Men's 200 Meters",
    "Men's 400 Meters", 
    "Men's 110 Meters Hurdles", 
    "Men's 400 Meters Hurdles", 
    "Men's 4x100 Meters Relay", 
    "Men's 4x400 Meters Relay", 
    "Women's 100 Meters", 
    "Women's 200 Meters", 
    "Women's 400 Meters", 
    "Women's 100 Meters Hurdles", 
    "Women's 400 Meters Hurdles", 
    "Women's 4x100 Meters Relay", 
    "Women's 4x400 Meters Relay", 
    "Mixed 4x400 Meters Relay"
]

country_specific_events_performance = events_performance[events_performance['Event'].isin(country_specific_events)]
country_specific_events_performance
```

```python
#Perform Couontry Ranking based Ranking Point
country_ranking = country_specific_events_performance.groupby("Country")["Ranking Point"].sum().reset_index()
country_ranking = country_ranking.sort_values(by = "Ranking Point", ascending= False)
country_ranking
```

```python
# Create Athlete Ranking and handle ties
country_ranking['Ranking'] = country_ranking['Ranking Point'].rank(method='min', ascending=False)
country_ranking['Ranking'] = country_ranking['Ranking'].astype(int)

# Use the Ranking column as index
country_ranking = country_ranking.set_index('Ranking')
country_ranking
```

```python
#Extract Top 20 Individual Athletes
top_20_athlete = athletes_ranking.head(20)

#Extract Top 10 Countries across the 15 Sprint, 
#Hurdles & Relay Events only - Women's & Men's 100m, 200m, 400m, 100/110mH, 400mH, 4x100m Relays, 4x400m Relays 
#AND Mixed 4x400m Relays 
top_10_country = country_ranking.head(10)
```
